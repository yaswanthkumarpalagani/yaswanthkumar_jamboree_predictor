```python
# Applying Min-Max scaling to normalize all the numerical columns in the DataFrame, and creating a new DataFrame with the scaled values.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1 = pd.DataFrame(scaler.fit_transform(df),columns = df.columns)
df1
```

```python
# Splitting the DataFrame into features (x) and target (y), where 'Chance of Admit' is the target variable, and checking the shapes of both x (features) and y (target).
from sklearn.model_selection import train_test_split
y = df1["Chance of Admit "]
x = df1.drop("Chance of Admit ",axis = 1)
y.shape,x.shape
```

```python
# Splitting the data into training and testing sets, with 30% of the data reserved for testing, and checking the shapes of the resulting train and test sets for features (x) and target (y).
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
# Importing necessary libraries for linear regression modeling: LinearRegression for standard linear regression, Lasso and Ridge for regularized regression, and mean_squared_error for evaluating model performance.
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.metrics import mean_squared_error
```

```python
# Initializing models for linear regression, Lasso regression (with alpha=0.1), and Ridge regression (with alpha=0.1) to apply to the data.
linear_model = LinearRegression()
lasso_model = Lasso(alpha = 0.1)
ridge_model = Ridge(alpha = 0.1)
```

```python
# Fitting the linear regression model to the training data (x_train, y_train) to learn the relationship between the features and the target variable.
linear_model.fit(x_train,y_train)
```

```python
# Fitting the Lasso regression model to the training data.
lasso_model.fit(x_train,y_train)
```

```python
# Fitting the Ridge regression model to the training data.
ridge_model.fit(x_train,y_train)
```

```python
# Printing the coefficients of the Linear, Lasso, and Ridge regression models to compare their learned weights for each feature.
print("Linear Regression --> ",linear_model.coef_)
print("Lasso Regression --> ",lasso_model.coef_)
print("Ridge Regression --> ",ridge_model.coef_)
```

```python
# Printing the intercepts (bias terms) of the Linear, Lasso, and Ridge regression models to compare their learned constant values.
print("Linear Regression --> ",linear_model.intercept_)
print("Lasso Regression --> ",lasso_model.intercept_)
print("Ridge Regression --> ",ridge_model.intercept_)
```

```python
# Printing the R-squared (coefficient of determination) scores for the Linear, Lasso, and Ridge regression models on the test data to evaluate their performance.
print("Linear Regression --> ",linear_model.score(x_test,y_test))
print("Lasso Regression --> ",lasso_model.score(x_test,y_test))
print("Ridge Regression --> ",ridge_model.score(x_test,y_test))
```

```python
# Calculating and printing the Mean Squared Error (MSE) for the Linear Regression model on both the training and test data to evaluate its prediction accuracy.
linear_train_mse = mean_squared_error(y_train,linear_model.predict(x_train))
print(linear_train_mse)
linear_train_mse = mean_squared_error(y_test,linear_model.predict(x_test))
print(linear_train_mse)
```

```python
# Calculating and printing the Mean Squared Error (MSE) for the Lasso Regression model on both the training and test data to evaluate its prediction accuracy.
lasso_train_mse = mean_squared_error(y_train,lasso_model.predict(x_train))
print(lasso_train_mse)
lasso_train_mse = mean_squared_error(y_test,lasso_model.predict(x_test))
print(lasso_train_mse)
```

```python
# Calculating and printing the Mean Squared Error (MSE) for the Ridge Regression model on both the training and test data to evaluate its prediction accuracy.
ridge_train_mse = mean_squared_error(y_train,ridge_model.predict(x_train))
print(ridge_train_mse)
ridge_train_mse = mean_squared_error(y_test,ridge_model.predict(x_test))
print(ridge_train_mse)
```
